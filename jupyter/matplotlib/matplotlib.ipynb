# matplotlib 基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y)
```

```python
plt.show()
```

```python
# 第二个图，绘制组合图片
cosy = np.cos(x)
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="-.") # 自定义颜色、样式
plt.show()
```

**line style**

- `:` ：点线
- `-.`：虚线线+点线
- `--`： 虚线
- `-`：直线

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.xlim(-2, 12)  # 控制x的最大、最小刻度
plt.ylim(-2, 2)  # 控制y的最大、最小刻度
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.axis((-2, 12, -2, 2)) # x/y的最大、最小刻度一起控制
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.xlabel("x axis") # x轴描述（中文存在异常）
plt.ylabel("y value")  # y轴描述
plt.show()
```

```python
plt.plot(x, siny, label="sin x value") # 两个走势的标签描述
plt.plot(x, cosy, label="cos y value")
plt.legend()  # 启用标签
plt.show()
```

```python
plt.plot(x, siny, label="sin x value")
plt.plot(x, cosy, label="cos y value")
plt.legend()
plt.title("sin and cos title") # 设置标题
plt.show()
```

## 散点图

```python
plt.scatter(x, siny)
plt.scatter(x, cosy)
plt.show()
```

```python
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)

plt.scatter(x, y, alpha=0.5) # 设置透明度
plt.show()
```
