```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def entropy(p):
    return -p * np.log(p) - (1-p) * np.log(1-p)
```

```python
entropy(0.3)
```

```python
x = np.linspace(0.01, 0.99, 200)
```

```python
plt.plot(x, entropy(x))
plt.show()
```

```python
# **说明：** 当 x 为 0.5 时，数据的不确定性是最明显的。当 x 为 0 或 1 时候，数据的确定性最强。
```
